```python
# reading in the Excel File

df1 = pd.read_excel(r"C:\Users\shero\Downloads\seasons(updated).xlsx",sheet_name='Sheet1')
df1
```

```python
# filling the NaN's with blanks

df1_fill = df1.fillna('')
df1_fill
```

```python
#setting the default index to Year

df1 = df1.set_index('Year')
df1
```

```python
#sorting values by Year in ascending order 

df1 = df1.sort_values(by = 'Year', ascending = True).head(25000)
df1
```

```python
# Removing the values that are below 25 games

df1 = df1[df1['G'] >= 25]
df1
```

```python
# Creating a sub group that includes point guards and shooting guards

guards = ['PG','SG']

df1[df1['Pos'].isin(guards)]
```

```python
# Creating a sub group that includes small forwards and power forwards


forwards = ['SF', 'PF']

df1[df1['Pos'].isin(forwards)]
```

```python
# Creating a sub group for centers

center = ['C']

df1[df1['Pos'].isin(center)]
```

```python
#filter/selecting specific columns

df1 = df1.filter(items = ['Player','Age','PPG', 'ASG', 'RBG','STL/Dis','BLK/Dis','FG/Dis','3Pper','2p/Dis','FT/Dis','Pos'])
df1
```
